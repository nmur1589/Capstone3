```python
# transform a time series dataset into a supervised learning dataset
# inputs are dataframe, your desired predictor variable, number of lags, and number of periods out to forecast

def series_to_supervised(data, yvar, n_in=1, n_out=1, dropnan=True):
    n_vars = 1 if type(data) is list else data.shape[1]
    df = pd.DataFrame(data)
    cols = list()
    colnames = list()
    # input sequence (t-n, ... t-1)
    for i in range(n_in, 0, -1):
        cols.append(df.shift(i))
        for n in df.columns:
            colnames.append(n + '-' + str(i))
            
            
    # forecast sequence (t, t+1, ... t+n)
    for i in range(0, n_out):
        
        cols.append(df[[yvar]].shift(-i))
        for n in df[[yvar]].columns:
            colnames.append(n + '+' + str(i))
        # put it all together
        agg = pd.concat(cols, axis=1)
        # drop rows with NaN values
        if dropnan:
            agg.dropna(inplace=True)
        agg.columns = colnames
    return agg
 


```

```python
#Time series train/test split function from the original project

def tscv(df, cv,test_size, min_samples = 60):
    
    
    #training size should not go below 60 records
    #divide training size by crossfolds desired
    #if less than 60 use 60, otherwise use teh amount
    train_size = round(len(df)/cv)
    if train_size > min_samples:
        train_size = round(len(df)/cv)
    else:
        train_size = 60
    
    
    
    #establish the first position to start
    
    first_train = len(df) - (train_size + ((test_size * (cv-1))) + test_size)
    
    train = []
    test = []
    
    #create the train/test splits
    
    n = 0
    
    for i in range(0,cv+ 1):

        if i == 0:
            train.append(df[0:first_train])
            test.append(df[first_train: (first_train + test_size)])
        else:
            train.append(df[first_train:(first_train+train_size)])
            test.append(df[(first_train + train_size): (first_train + train_size + test_size)])

            first_train = first_train + test_size

    return train, test
   
    
```

```python
#function to visualize what the splits look like

def splits(train, test):

    tstart = []
    tend = []
    test_start = []
    test_end = []
    for n in range(0, len(train)):

        tstart.append(train[n][0:1].index[0])
        tend.append(train[n][-1:].index[0])
        test_start.append(test[n][0:1].index[0])
        test_end.append(test[n][-1:].index[0])
    
    return pd.DataFrame({'train_from':tstart, 'train_to':tend, 'test_from':test_start, 'test_to': test_end}, index = range(0,len(train)))
```

```python
# dual axis plot function
#I will be plotting RMSE and R2 metrics for each forecasted period in the evaluation section

import matplotlib.pyplot as plt
def dual_axis(x, y1,y2, y1_title, y2_title):

    fig, ax1 = plt.subplots()
    t = x
    data1 = y1
    data2 = y2

    color = 'tab:red'
    ax1.set_xlabel('Month')
    ax1.set_ylabel(y1_title, color=color)
    ax1.plot(t, data1, color=color)
    ax1.tick_params(axis='y', labelcolor=color)

    ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

    color = 'tab:blue'
    ax2.set_ylabel(y2_title, color=color)  # we already handled the x-label with ax1
    ax2.plot(t, data2, color=color)
    ax2.tick_params(axis='y', labelcolor=color)

    fig.tight_layout()  # otherwise the right y-label is slightly clipped
    
    plt.show()
```

```python
#Time series regressor function

def tsRegressor(train, test, model, rt = 'all'):


    lst_accuracy = []

    for n in range(0, len(train)):

        predictions = []
        #split x and y variables
        
        if rt == 'all':
            
            y = train[n].iloc[:,-12:]
        else:
            y = train[n].iloc[:,-1:]
            
        X = train[n].iloc[:,:-12]

        #set test index = to first row in test array
        #this will include all lagged values for X and exclude the y pricing variables
        testindex = test[n].iloc[0:1].index.values[0]
        Xtest_n = tsdf[testindex:testindex].iloc[:,:-12]

        #create a step counter to identify the step in the forecast
        # i.e. step 0 is forecasting 1 month out, step 1 forecasting 2 months out.. and so on until 12 months
        step = 0
        steplist = []

        #loop through each y value(where y = t+0, t+1, t+2, etc)
        for cols in y:
            #model = XGBRegressor(objective='reg:squarederror', n_estimators=1000)
            model.fit(X.values, y[cols])
            ypred = model.predict(Xtest_n.values.reshape(1,-1))
            steplist.append(step)
            predictions.append(ypred[0])
            step = step + 1


        #create list of actual values 
        datefrom = pd.to_datetime(testindex)
        
        dateto = datefrom + MonthEnd(11)
        
        if rt == 'all':
            ytest = df2[datefrom:dateto].Price
        else:
            ytest = df2[dateto:dateto].Price

        #create accuracy dataframe
        accuracy = pd.DataFrame({'Actuals': ytest, 'Forecast': predictions, 'Step': steplist})
        lst_accuracy.append(accuracy)

    return lst_accuracy
```

```python
c2results = tsRegressor(train,test, XGBRegressor(objective='reg:squarederror', n_estimators=100), rt = 'last')
```

```python
#store c2 results

c2results = tsRegressor(train,test, XGBRegressor(objective='reg:squarederror', n_estimators=100))
```

```python
def acc_by_step(results):
    lstrmse = []
    lstr2 = []
    fullbacktest = pd.concat(results)


    for i in range(0,12):
        steptest = fullbacktest[fullbacktest.Step == i]
        lstrmse.append(np.sqrt(mean_squared_error(steptest.Actuals, steptest.Forecast)))
        lstr2.append(r2_score(steptest.Actuals, steptest.Forecast))
        
    return pd.DataFrame({'RMSE': lstrmse, 'r2': lstr2})
```

### Tune estimator parameter

```python
estimators = [10, 50, 100, 500, 1000]
lstrmse = []
for n in estimators:
    paramtuning = tsRegressor(train, test,XGBRegressor(objective='reg:squarederror', n_estimators=n))
    lstrmse.append(np.mean(acc_by_step(paramtuning).RMSE))
    
```
